```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
df=pd.read_csv("/content/drive/MyDrive/Chaithra/BL-Flickr-Images-Book.csv")
print(df.head())
```

```python
to_drop=["Edition Statement","Corporate Author","Former owner"]
df.drop(to_drop,inplace=True,axis=1)
df.head()
```

```python
df.shape
```

```python
for col in df.columns:
  print(col)
```

```python
df.set_index("Identifier",inplace=True)
df.head(15)
```

```python
unwanted_characters=['[',',','-']
def clean_dates(item):
  dop=str(item.loc['Date of Publication'])
  if dop =='nan' or dop[0] =='[':
    return np.NaN
  for character in unwanted_characters:
    if character in dop:
      character_index=dop.find(character)
      dop=dop[:character_index]
  return dop
df['Date of Publication']=df.apply(clean_dates,axis=1)
```

```python
df.head(20)
```

```python
def clean_title(title):
  if title =='nan':
    return 'NaN'
  if title[0] =='[':
    title=title[1:title.find(']')]
  if 'by' in title:
    title=title[:title.find('by')]
  elif 'By' in title:
    title=title[:title.find('By')]
  if '[' in title:
    title=title[:title.find('[')]
  title=title[:-2]
  return title

df['Title']=df['Title'].apply(clean_title)
df.head()
```
